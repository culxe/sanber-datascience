**Soal 4. Time grouping**

 

*   Lakukan grouping data perubahan iklim pada tahun 2016 berdasarkan frekuensi 1 bulan dengan memberikan agregasi max pada kolom meantemp. humidity, wind_speed dan meanpressure!
*   Berikan analisa, di bulan apa rata-rata temperatur, humidity, wind_speed dan meanpressure tertinggi?





```python
#code here
weekly_agg = df.loc['2016'].groupby(pd.Grouper(freq='1W')).agg({
    'meantemp': 'first',  
    'humidity': 'max',    
    'wind_speed': 'min',  
    'meanpressure': 'last',
    'meanpressure': 'sum'  
})
weekly_agg.head(10)
```

```python
monthly_avg = df.loc['2016'].resample('M').mean()

print('Di bulan',monthly_avg['meantemp'].idxmax().month, 'didapatkan meantemp tertinggi')
print('Di bulan',monthly_avg['humidity'].idxmax().month,'didapatkan humidity tertinggi')
print('Di bulan',monthly_avg['wind_speed'].idxmax().month,'didapatkan wind speed tertinggi')
print('Di bulan',monthly_avg['meanpressure'].idxmax().month,'didapatkan mean pressure tertinggi')
```
