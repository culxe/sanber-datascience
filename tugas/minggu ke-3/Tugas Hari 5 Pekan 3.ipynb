```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python

df = sns.load_dataset('diamonds')
```

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

```python
sns.set({'figure.figsize':(13,13)})
palette = "ch:r=-.2,d=.3_r"
 
#code here
sns.scatterplot(data=df, x='carat', y='price', hue='cut', palette=palette)
plt.show()
```

# Soal 2: Catplot

Buatlah sebuah Categorical plot dengan kind = 'boxplot', sehingga diperoleh hasil seperti expected output berikut:


```python
#code here

sns.catplot(
    data=df,
    x='clarity',
    y='price',
    col='cut',
    kind='box',
    height=4,
    aspect=1
)
plt.show()
```

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price serta markers =  ['p', 'o', 's', 'D', 'X'] sehingga dihasilkan output sesuai expected.

```python
df = df.sample(n=300, random_state=123)

#code here
sns.pairplot(
    data=df,
    vars=['carat', 'depth', 'table', 'price'],
    hue='cut',  # Menggunakan kolom 'cut' untuk kategori
    markers=['p', 'o', 's', 'D', 'X']  # Marker spesifik untuk kategori
)

# Menampilkan plot
plt.show()
```

# Soal 4: Joint Plot

Gunakan data yang sama dengan soal 3 untuk menghasilkan dua visualisasi joint plot antara carat dan price nya sesuai expected output.

```python
#code here
with sns.axes_style('white'):
    g1 = sns.jointplot(
        data=df,
        x="carat",
        y="price",
        kind="kde",
        fill=False,
        color="blue"
    )
    g1.fig.suptitle("KDE Jointplot", y=1.02)  
    plt.show()
with sns.axes_style('white'):
    g2 = sns.jointplot(
        data=df,
        x="carat",
        y="price",
        kind="reg",
        scatter_kws={'alpha': 0.6},
        line_kws={"color": "blue"}
    )
    g2.fig.suptitle("Scatterplot with Regression", y=1.02)  
    plt.show()
```
