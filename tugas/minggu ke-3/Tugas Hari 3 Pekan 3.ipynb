<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2*-np.pi, 2*np.pi, 200)

tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)

```

```python
#code here
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(12,8), nrows=1, ncols=3, sharex=True, sharey=True)

for axes in ax:
    axes.set_facecolor('white')
    axes.tick_params(axis='both', colors='black')


ax[0].plot(x, tan, color='Blue',  label='tanX')
ax[1].plot(x, cos, color='Red', label='cosX')
ax[2].plot(x, sin, color='Green', label='sinX')

fig.legend(loc='upper center')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(x, y, color='steelblue')
ax.set_facecolor('white')
ax.tick_params(axis='both', colors='black')


ins1 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
ins1.plot(x, y, color='blue')
ins1.plot(x, y2, color='gold', label='Grafik Cos X^2')
ins1.set_facecolor('white')
ins1.tick_params(axis='both', colors='black')
ins1.set_title('Grafik Cos X^2', fontsize=10, color='black')

ins2 = ax.inset_axes([-0.1, 0.9, 0.3, 0.3])
ins2.plot(x, y, color='blue')
ins2.plot(x, y3, color='black')
ins2.set_facecolor('white')
ins2.tick_params(axis='both', colors='black')
ins2.set_title('Grafik Cos X^3', fontsize=10, color='black')

ins3 = ax.inset_axes([-0.1, -0.1, 0.3, 0.3])
ins3.plot(x, y, color='blue')
ins3.plot(x, y3, color='red')
ins3.set_facecolor('white')
ins3.tick_params(axis='both', colors='black')
ins3.set_title('Grafik Cos X^4', fontsize=10, color='black')

ins4 = ax.inset_axes([0.8, -0.1, 0.3, 0.3])
ins4.plot(x, y, color='blue')
ins4.plot(x, y4, color='green')
ins4.set_facecolor('white')
ins4.tick_params(axis='both', colors='black')
ins4.set_title('Grafik Cos X^5', fontsize=10, color='black')

plt.show()
```
